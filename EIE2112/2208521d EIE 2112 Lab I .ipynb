```python
import matplotlib.pyplot as plt
import numpy as np
#define func
f = lambda x: x**2*np.cos(x)- x
#define f'(x)
def derivative(f, x):
  delta_x = 1/1000000
  return (f(x+delta_x)-f(x))/delta_x
#define gradient descent algorithm
def find_min_gradient(start_point, step, num_iterations):
  x = start_point
  #list store x
  x_history = [x]
  for i in range(num_iterations):
    #update rule
    grad_x = derivative(f, x)
    x = x - step*grad_x
    x_history.append(x) #store each x
    #calculted gradient for updated x
    new_grad_x = derivative(f, x)
    #stop criteria: new_grad_x = 0 or new_grad_x changes sign
    if new_grad_x == 0 or new_grad_x*grad_x < 0:
      print("The local minimum point at x=", x, "y=", f(x))
      break
  return x_history

x_history = find_min_gradient(start_point=-0.05, step= 0.1, num_iterations=200)
x_range = np.linspace(-5, 5, 100)
plt.plot(x_range, f(x_range), label="f(x)")
#tranfer list to array for x_history
plt.plot(x_history, f(np.array(x_history)), 'rx', label='update x')
plt.legend(loc="best")
plt.show()
```
