```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
# from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

data = pd.read_csv('Advertising.csv')
print(data.head())
print('Dataset dimension:', data.shape)

TV = data['TV']
Radio = data['Radio']
Newspaper = data['Newspaper']
Sales = data['Sales']

plt.subplot(1, 3, 1)
plt.scatter(TV, Sales, color='red')
plt.xlabel("TV")
plt.ylabel("Sales")

plt.subplot(1, 3, 2)
plt.scatter(Radio, Sales, color='red')
plt.xlabel("Radio")

plt.subplot(1, 3, 3)
plt.scatter(Newspaper, Sales, color='red')
plt.xlabel("Newspaper")

plt.show()

linear = linear_model.LinearRegression()
TV = TV.values.reshape(-1,1)
linear.fit(TV, Sales)

print('Coefficients:', linear.coef_)
print('Intercept:', linear.intercept_)

predictions = linear.predict(TV)
plt.scatter(TV, Sales, color='red')
plt.plot(TV, predictions, color='blue', linewidth=3)
plt.xlabel("TV")
plt.ylabel("Sales")
plt.show()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

Iris = pd.read_csv('Iris.csv')
print(Iris.head())
print('Dataset dimension:', Iris.shape)

Iris.columns =['sep_L', 'sep_W', 'pet_L', 'pet_W', 'Type']
print(Iris.head())

X = Iris[['sep_L', 'sep_W', 'pet_L', 'pet_W']]
y = Iris['Type']

# np.random.seed(0)
(x_train, x_test, y_train, y_test) = train_test_split(X, y, test_size=0.3)

dt = tree.DecisionTreeClassifier()
dt = dt.fit(x_train, y_train)
tree.plot_tree(dt)

train_acc = dt.predict(x_train)
test_acc = dt.predict(x_test)
print('Training accuracy:', accuracy_score(y_train, train_acc))
print('Test accuracy:', accuracy_score(y_test, test_acc))
```

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Activation, Flatten, Dropout
from keras.optimizers import SGD
import matplotlib.pyplot as plt
import os

(x_train, y_train), (x_test, y_test) = mnist.load_data() #load mnist dataset
x_train = x_train.astype('float32') #more reshaping
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255

num_category = 10   #set number of categories
y_train = keras.utils.to_categorical(y_train, num_category) #convert class vectors to binary class matrices
y_test = keras.utils.to_categorical(y_test, num_category)

print('Train: X=%s, y=%s' % (x_train.shape, y_train.shape))
print('Test: X=%s, y=%s' % (y_test.shape, y_test.shape))

# plot first few images
for i in range(9):
  plt.subplot(330 + 1 + i)
  plt.imshow(x_train[i], cmap=plt.get_cmap('gray'))
plt.show()
```

```python
## FULLY CONNECTED NEURAL NETWORK
image_vector_size = 28*28
x_train_fcnn = x_train.reshape(x_train.shape[0], image_vector_size)  # Flatten the images
x_test_fcnn = x_test.reshape(x_test.shape[0], image_vector_size)

model = Sequential()
model.add(Dense(units=32, activation='sigmoid', input_shape=(image_vector_size,)))
model.add(Dense(units=num_category, activation='softmax'))

model.summary()
model.compile(optimizer="sgd", loss='categorical_crossentropy', metrics=['accuracy'])
fcnn_model_log = model.fit(x_train_fcnn, y_train,
                           batch_size=128,
                           epochs=10,
                           verbose=True,
                           validation_split=.1)
fcnn_loss, fcnn_accuracy  = model.evaluate(x_test_fcnn, y_test, verbose=False)
# print(f'Test loss: {fcnn_loss:.3}')
print(f'Test accuracy: {fcnn_accuracy:.3}')

fig = plt.figure()
plt.plot(fcnn_model_log.history['accuracy'])
plt.plot(fcnn_model_log.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()
```

```python
## CONVOLUTIONAL NEURAL NETWORK

#For 3D data, "channels_last" assumes (conv_dim1, conv_dim2, conv_dim3, channels) while
#"channels_first" assumes (channels, conv_dim1, conv_dim2, conv_dim3).
img_rows = 28
img_cols = 28
x_train_cnn = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)  # Reshaping
x_test_cnn = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)

model = Sequential() #convolutional layer with rectified linear unit activation
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=input_shape)) #32 convolution filters used each of size 3x3
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten()) #flatten since too many dimensions, we only want a classification output
model.add(Dense(100, activation='relu')) #fully connected to get all relevant data
model.add(Dense(num_category, activation='softmax')) #output a softmax to squash the matrix into output probabilities

model.summary()
opt = SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

##model training
cnn_model_log = model.fit(x_train_cnn, y_train,
                      batch_size=128,
                      epochs=10,
                      verbose=1,
                      validation_split=.1)

cnn_loss, cnn_accuracy  = model.evaluate(x_test_cnn, y_test, verbose=0)
print(f'CNN test accuracy: {cnn_accuracy:.3}')

# plotting the metrics
fig = plt.figure()
plt.plot(cnn_model_log.history['accuracy'])
plt.plot(cnn_model_log.history['val_accuracy'])
plt.title('CNN Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='lower right')
```

```python
from keras.datasets import cifar10
(trainX, trainY), (testX, testY) = cifar10.load_data()
trainX.shape

## CONVOLUTIONAL NEURAL NETWORK

#For 3D data, "channels_last" assumes (conv_dim1, conv_dim2, conv_dim3, channels) while
#"channels_first" assumes (channels, conv_dim1, conv_dim2, conv_dim3).
img_rows = 32
img_cols = 32
trainX_cnn = trainX.reshape(trainX.shape[0], img_rows, img_cols, 3)  # Reshaping
testX_cnn = testX.reshape(testX.shape[0], img_rows, img_cols, 3)
input_shape = (img_rows, img_cols, 3)

model = Sequential() #convolutional layer with rectified linear unit activation
model.add(Conv2D(50, kernel_size=(3,3), activation='relu', input_shape=input_shape)) #32 convolution filters used each of size 3x3
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten()) #flatten since too many dimensions, we only want a classification output
model.add(Dense(100, activation='relu')) #fully connected to get all relevant data
model.add(Dense(num_category, activation='softmax')) #output a softmax to squash the matrix into output probabilities

model.summary()
opt = SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

# One-hot encode the labels
from tensorflow.keras.utils import to_categorical
trainY = to_categorical(trainY, num_classes=num_category)  # Assuming 'num_category' is defined
testY = to_categorical(testY, num_classes=num_category)

##model training
cnn_model_log = model.fit(trainX_cnn, trainY,
                      batch_size=128,
                      epochs=10,
                      verbose=1,
                      validation_split=.1)

cnn_loss, cnn_accuracy  = model.evaluate(testX_cnn, testY, verbose=0)
print(f'CNN test accuracy: {cnn_accuracy:.3}')

# plotting the metrics
fig = plt.figure()
plt.plot(cnn_model_log.history['accuracy'])
plt.plot(cnn_model_log.history['val_accuracy'])
plt.title('CNN Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='lower right')
```
